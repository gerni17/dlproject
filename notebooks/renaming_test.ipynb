```python
path="/home/gian/git/deeplabv3p/data/exp/train"
for dirname in os.listdir(path):
    if os.path.isdir(os.path.join(path, dirname)):
        print("it is")
        path_dir=os.path.join(path, dirname)
        files=[name for name in os.listdir(path_dir) if os.path.isfile(os.path.join(path_dir, name))]
        print(files)
        i=0
        for i, filename in enumerate(files):
            print(dirname + "/" + filename)
            print(os.path.join(path, dirname) + "/" + str(i) + ".png")
            os.rename(os.path.join(path, dirname) + "/" + filename, os.path.join(path, dirname) + "/" + str(i) + ".png")
    else:
        print(dirname," not a directory")
```
